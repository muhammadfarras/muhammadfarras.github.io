## Function

Fungsi adalah sebuah ekspresi yang mendifinisikan hubungan antara dua atau lebih variable. Fungsi membutuhkan _independent variable_ atau _domain variable_ dan menghasilkan _dependent variable_ atau _features_.

Dibawah ini adalah contoh dari sebuah function

_f(x) = 2x + 1_

```python
## Fungsi pada python
def f (x):
    return (2*x) + 1
```

```python
## Variable independent
dataIndependent = np.arange(10)
dataDependent = f(dataIndependent)

dataIndependent, dataDependent
```

```python
dataFrameData = {"Independent":dataIndependent, "Dependent":dataDependent}
pd.DataFrame(dataFrameData)
```

Selanjutnya kita akan memplot data diatas menggunakan `Matplotlib` untuk melihat plot dari **Linear function** _f(x) = 2x +1_

```python
from sympy import *

x = symbols('x')
f = 2*x + 1
plot (f)
```

**Curvilinear Functions** adalah sebuah fungsi yang ketika di plot akan membuat sebuah cekungan.

```python
x = symbols('x')
f = x**2 + 4*x - 4
plot (f)
```

## Logarithms

Adalah sebuah fungsi matematika untuk menentukan kuadrat dan nilai dasar dari sebuah nilai.





```python
value  = log(8,2)
print(value)
```

Jika kita tidak memberikan argument angka dasar, pada contoh diatas adalah angka **2** pada fungsi `log(8,2)`. Maka secara otomatis menggunakan nilai bawaan, yaitu 10.

Namun pada data science, nilai dasar bawaan untuk log adalah _Euler's number_ **_e_**.

Sama seperti kuadrat (exponent), Logarithms juga dapat dikalikan, dibagi, ditambah dan dikurangi. Dibawah ini adalah table cara melakukan operasi matematika pada exponent dan logarithms



```python
## Multiplocation Logarithm property
print(log(4,2)+log(4,2))
print(log((4*4),2))
```
